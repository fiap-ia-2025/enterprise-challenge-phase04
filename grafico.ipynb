```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('medicao_vibracao.csv')
```

```python
# Mapeamento de cores
cor_mapa = {
    'NORMAL': 'green',
    'ALERTA_Pre_falha': 'orange',
    'FALHA_CRITICA': 'red'
}
cores = df['Status'].map(cor_mapa)

# Gráfico de barras
plt.figure(figsize=(16, 6))
plt.bar(df['ID'], df['Aceleracao_Total'], color=cores, width=0.5)

plt.xlabel('Medições')
plt.ylabel('Aceleração Total')
plt.title('Aceleração por status ao longo das medições')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Legenda
import matplotlib.patches as mpatches
legenda = [mpatches.Patch(color=v, label=k) for k, v in cor_mapa.items()]
plt.legend(handles=legenda)

plt.tight_layout()
plt.show()
```
